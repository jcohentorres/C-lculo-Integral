```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad



# Definir las funciones
def f(x):
    return x**2

def g(x):
    return x

# Definir el intervalo de integración
a = 0  # límite inferior
b = 1  # límite superior

# Calcular el área entre las curvas (integral de la diferencia entre las funciones)
# En este caso, área entre f(x) y g(x) en el intervalo [a, b]
area, error = quad(lambda x: -f(x) + g(x), a, b)

# Imprimir el área calculada
print(f"Área entre las curvas: {area:.3f} unidades cuadradas")

# Crear un conjunto de puntos para graficar las funciones
x_vals = np.linspace(a, b, 500)
y_f = f(x_vals)
y_g = g(x_vals)

# Graficar las funciones
plt.plot(x_vals, y_f, label=r'$f(x) = x^2$', color='orange')
plt.plot(x_vals, y_g, label=r'$g(x) = x$', color='black')

# Rellenar el área entre las curvas
plt.fill_between(x_vals, y_f, y_g, where=(y_f > y_g), color='lightblue', alpha=0.5, label="Área entre curvas")

# Etiquetas y leyenda
plt.title("Área entre las curvas")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definimos la función que queremos integrar
def f(x):
    return np.exp(x)  # Puedes cambiar esta función por cualquier otra

# Límites de integración
a = 0  # Límite inferior
b = np.pi  # Límite superior

# Calculamos el área bajo la curva usando la integral definida
area, error = quad(f, a, b)

# Imprimimos el área calculada
print(f"El área bajo la curva es: {area:.4f}")

# Graficamos la curva y el área sombreada
x = np.linspace(a, b, 500)
y = f(x)

plt.figure(figsize=(8, 6))

# Graficamos la curva
plt.plot(x, y, label=r'$f(x) = \exp(x)$', color='black')

# Sombreando el área debajo de la curva
plt.fill_between(x, y, where=(y >= 0), color='red', alpha=0.5, label='Área bajo la curva')

# Añadimos etiquetas y leyenda
plt.title("Área bajo la curva de $f(x) = \sin(x)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
