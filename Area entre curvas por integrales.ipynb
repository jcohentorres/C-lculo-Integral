```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad



# Definir las funciones
def f(x):
    return x**2

def g(x):
    return x

# Definir el intervalo de integraci√≥n
a = 0  # l√≠mite inferior
b = 1  # l√≠mite superior

# Calcular el √°rea entre las curvas (integral de la diferencia entre las funciones)
# En este caso, √°rea entre f(x) y g(x) en el intervalo [a, b]
area, error = quad(lambda x: -f(x) + g(x), a, b)

# Imprimir el √°rea calculada
print(f"√Årea entre las curvas: {area:.3f} unidades cuadradas")

# Crear un conjunto de puntos para graficar las funciones
x_vals = np.linspace(a, b, 500)
y_f = f(x_vals)
y_g = g(x_vals)

# Graficar las funciones
plt.plot(x_vals, y_f, label=r'$f(x) = x^2$', color='orange')
plt.plot(x_vals, y_g, label=r'$g(x) = x$', color='black')

# Rellenar el √°rea entre las curvas
plt.fill_between(x_vals, y_f, y_g, where=(y_f > y_g), color='lightblue', alpha=0.5, label="√Årea entre curvas")

# Etiquetas y leyenda
plt.title("√Årea entre las curvas")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Mostrar el gr√°fico
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definimos la funci√≥n que queremos integrar
def f(x):
    return np.exp(x)  # Puedes cambiar esta funci√≥n por cualquier otra

# L√≠mites de integraci√≥n
a = 0  # L√≠mite inferior
b = np.pi  # L√≠mite superior

# Calculamos el √°rea bajo la curva usando la integral definida
area, error = quad(f, a, b)

# Imprimimos el √°rea calculada
print(f"El √°rea bajo la curva es: {area:.4f}")

# Graficamos la curva y el √°rea sombreada
x = np.linspace(a, b, 500)
y = f(x)

plt.figure(figsize=(8, 6))

# Graficamos la curva
plt.plot(x, y, label=r'$f(x) = \exp(x)$', color='black')

# Sombreando el √°rea debajo de la curva
plt.fill_between(x, y, where=(y >= 0), color='red', alpha=0.5, label='√Årea bajo la curva')

# A√±adimos etiquetas y leyenda
plt.title("√Årea bajo la curva de $f(x) = \sin(x)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```
